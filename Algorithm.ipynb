```python
# 최대값을 찾는 알고리즘

# 첨자 index   0  1  2  3  4
# 데이터      12 13 11 14 10

arr = [12, 13, 11, 14, 10]

max = arr[0]

for i in arr :
  if i > max :
    max = i
    i = i + 1
  else :
    i = i + 1
print(max)

#반복문엔 탈출지점 필수!
```

```python
# Linear Search 선형 탐색법

# 보관된 데이터를 맨 앞부터 순서대로 원하는 값을 찾을 때까지 탐색한다.
# 탐색 처리는 반복 구조를 사용하여 기술한다.
# 반복 구조에서는 종료 조건을 잊지 말자!

# 첨자 (방번호) 0 1 2 3 4
# 데이터        4 2 3 5 1    찾을 데이터 : 5

arr = [4,2,3,5,1]

for i in range(0,5) : #레인지로 하면 인덱스 번호  #레인지 안써서 한번 만들어보기
  if arr[i] == 5 : # 그래서 어레이의 몇번째로
    print(f'{i}번째 요소가 일치')
    break
  elif i == 4:
    print('찾지 못했습니다')


# for 문이라 자기가 알아서 하나씩 증가하기 때문에 i = i + 1 필요가 없음
# 0 -> i , i = i + 1 , i < len(arr) 알고리즘이 for문으로 다 처리가 가능
```

```python
def linear_search(no, list) :
  for i in range(len(list)) :
    if no == list[i] :
      return print(f'{i}번째 요소가 일치')
  return print('찾지 못했습니다') # -1 또는 None

# 왜 계속 none이 나오는가

print(linear_search(5, [4, 2, 3, 5, 1]))
```
